**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
#sobel in x Richtung (horizontal)
def sobel_h(img):
    img_sobel_h = np.zeros_like(img)     #gleich groß wie img -> leerer array
    for i in range(1,img.shape[0]-1):
        for j in range(1,img.shape[1]-1):
            img_sobel_h[i, j] = 1/8*(img[i-1][j-1] + img[i][j-1]*2 + img[i+1][j-1] - img[i-1][j+1] - 2*img[i][j+1] - img[i+1][j+1])

    return img_sobel_h
```

```python
#sobel in y Richtung (vertikal)
def sobel_v(img):
    img_sobel_v = np.zeros_like(img)     #gleich groß wie img -> leerer array
    for i in range(1, img.shape[0]-1):
        for j in range(1, img.shape[1]-1):
            img_sobel_v[i, j] = 1/8* (img[i-1][j-1] + img[i-1][j]*2 + img[i-1][j+1] - img[i+1][j-1] - 2*img[i+1][j] - img[i+1][j+1])
    return img_sobel_v
```

```python
##das gleiche OHNE loops -> umständlicher??

def sobel_h(img):
    # Sobel-Kernel für horizontale Kanten
    sobel_kernel = np.array([
        [1,  2,  1],
        [0,  0,  0],
        [-1, -2, -1]
    ]) / 8.0

    # Initialisiere ein leeres Array für das Ergebnis
    img_sobel_h = np.zeros_like(img, dtype=float)

    # Faltung mit NumPy-Array-Slicing
    img_sobel_h[1:-1, 1:-1] = (
        sobel_kernel[0, 0] * img[:-2, :-2] + sobel_kernel[0, 1] * img[:-2, 1:-1] + sobel_kernel[0, 2] * img[:-2, 2:] +
        sobel_kernel[1, 0] * img[1:-1, :-2] + sobel_kernel[1, 1] * img[1:-1, 1:-1] + sobel_kernel[1, 2] * img[1:-1, 2:] +
        sobel_kernel[2, 0] * img[2:, :-2] + sobel_kernel[2, 1] * img[2:, 1:-1] + sobel_kernel[2, 2] * img[2:, 2:]
    )

    return img_sobel_h
```
